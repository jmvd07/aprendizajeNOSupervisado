### Analisis de los datos de *MovieLens*

```python
import pandas as pd
import numpy as np
import sys
from itertools import combinations, groupby
from collections import Counter
from IPython.display import display
```

```python
def size(obj):
    return "{0:.2f} MB".format(sys.getsizeof(obj) / (1000 * 1000))
```

```python
ratings = pd.read_csv('ratings.csv')
# es muy pesado trabajar con toda la base...
print('ratings -- dimensions: {0};   size: {1}'.format(ratings.shape, size(ratings)))
display(ratings.head())
#display(ratings.tail())
movies = pd.read_csv('movies.csv')
display(movies.head())

#decodificar el nombre de las peliculas
movies_df = pd.merge(ratings[['userId', 'movieId']], movies[['movieId','title']] ,on='movieId', how= "inner")

display(movies_df.head())
movies_df = movies_df.sort_values( by='userId', axis=0, ascending=True, 
                                  inplace=False, kind='quicksort', na_position='last')
movies = movies_df.values[:,[0,2]]
print(movies)
```

```python
transactions=[]
for user_id, user_movie in groupby(movies, lambda x: x[0]):
    transactions.append([item[1] for item in user_movie])
#print(transactions)
#con toda la base..
#IOPub data rate exceeded.
#The notebook server will temporarily stop sending output
```

```python
from efficient_apriori import apriori
#min_support y min_confiden entre 0 y 1  
itemsets, rules = apriori(transactions, min_support=0.2,  min_confidence=0.4, max_length=4)
```
